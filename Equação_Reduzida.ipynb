#*Equação reduzida da circunferência*

```python
import matplotlib.pyplot as plt
import numpy as np

# Defina as coordenadas do centro (h, k) e o raio (r) da circunferência
h = 2
k = 3
r = 4

# Crie um intervalo de valores para x
theta = np.linspace(0, 2*np.pi, 100)

# Calcule as coordenadas (x, y) da circunferência
x = h + r * np.cos(theta)
y = k + r * np.sin(theta)

# Plote a circunferência
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Circunferência: $(x - 2)^2 + (y - 3)^2 = 4^2$')
plt.scatter(h, k, color='red', marker='o', label='Centro (h, k)')

# Adicione rótulos e legendas
plt.title('Circunferência no Plano Cartesiano')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Mostre o gráfico
plt.show()
```
